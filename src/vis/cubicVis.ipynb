```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plotCubicPlane(nx:int, ny:int, nz:int, path:str, title:str, x = None, y = None, z = None):
    data = np.fromfile(path, dtype=np.float32)
    data = data.reshape((nz, ny, nx))
    if x is not None:
        plt.imshow(data[:,:,x], cmap="viridis")
    elif y is not None:
        plt.imshow(data[:,y,:], cmap="viridis")
    elif z is not None:
        plt.imshow(data[z,:,:], cmap="viridis")
    else:
        raise ValueError("one of parameter x, y, z should not be None")
    plt.colorbar()
    plt.title(title)
    plt.show()
```

```python
nx =512
ny =256
nz =256
N = 20
pathPrefix = "/home/xusj/GSY/GsyFluid/src/build/example/D3Q27DoubleDDFSimulator/"
```

```python
for i in range(N):
    plotCubicPlane(nx, ny, nz, pathPrefix+f"frame_{i}.dat", f"vx_frame_{i}", y = ny//2)
```
