```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plotCubicPlane(nx:int, ny:int, nz:int, path:str, title:str, x = None, y = None, z = None, vmin = None, vmax = None):
    data = np.fromfile(path, dtype=np.float32)
    data = data.reshape((nz, ny, nx))
    if x is not None:
        plt.imshow(data[:,:,x], cmap="viridis", vmin=vmin, vmax=vmax)
    elif y is not None:
        plt.imshow(data[:,y,:], cmap="viridis", vmin=vmin, vmax=vmax)
    elif z is not None:
        plt.imshow(data[z,:,:], cmap="viridis", vmin=vmin, vmax=vmax)
    else:
        raise ValueError("one of parameter x, y, z should not be None")
    plt.colorbar()
    plt.title(title)
    plt.show()
```
